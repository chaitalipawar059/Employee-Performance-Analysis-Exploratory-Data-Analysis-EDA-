```python
import numpy as np
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv("C:\\Users\\hp\\Downloads\\Test_data (2).csv")
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor 
from statsmodels.tools import add_constant
```

```python
numeric_df = df.select_dtypes(include=["number"])
numeric_df.reset_index(drop=True, inplace = True)
numeric_df.corr().round(2)
```

```python
corr_matrix = numeric_df.corr()
corr_matrix["KPIs_met >80%"].sort_values(ascending=False)
```

```python

features = ["previous_year_rating", "awards_won?","avg_training_score", "no_of_trainings","length_of_service"]
X = numeric_df[features]
X
```

```python
X_const = add_constant(X)
X_const
```

```python
vif_data = pd.DataFrame()
vif_data["feature"] = X.columns
vif_data["VIF"] = [variance_inflation_factor(X_const.values, i+1) for i in range(len(X.columns))]
vif_data
```

```python
y = numeric_df["KPIs_met >80%"]
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
# split the data in training and test sample 
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.33, random_state=42)
```

```python
len(X_train), len(X_test), len(y_train), len(y_test)
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
model.intercept_, model.coef_.round(2)
```

```python
# model valuation 
y_pred = model.predict = model.predict(X_test)
```

```python
y_test, y_pred
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, root_mean_squared_error
```

```python
mae = mean_absolute_error(y_test , y_pred)
mse = mean_squared_error(y_test , y_pred)
r_sq_score = r2_score(y_test, y_pred)
rmse = root_mean_squared_error(y_test, y_pred)
```

```python
print(f"mae:{mae}\nmse:{mse}:\nrmse:{rmse}\nr2_score:{r_sq_score}")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

# XGBoost Baseline Model

```python
from xgboost import XGBClassifier

xgb = XGBClassifier(
    n_estimators=200,
    learning_rate=0.05,
    max_depth=6,
    random_state=42,
    eval_metric="logloss"
)

xgb.fit(X_train, y_train)
xgb_pred = xgb.predict(X_test)

xgb_acc = accuracy_score(y_test, xgb_pred)
print("XGBoost Accuracy:", round(xgb_acc, 4))
```

```python
from catboost import CatBoostClassifier

cat = CatBoostClassifier(
    iterations=300,
    learning_rate=0.05,
    depth=6,
    verbose=False,
    random_seed=42
)

cat.fit(X_train, y_train)
cat_pred = cat.predict(X_test)

cat_acc = accuracy_score(y_test, cat_pred)
print("CatBoost Accuracy:", round(cat_acc, 4))
```

```python
from lightgbm import LGBMClassifier

lgb = LGBMClassifier(
    n_estimators=300,
    learning_rate=0.05,
    max_depth=-1,
    random_state=42
)

lgb.fit(X_train, y_train)
lgb_pred = lgb.predict(X_test)

lgb_acc = accuracy_score(y_test, lgb_pred)
print("LightGBM Accuracy:", round(lgb_acc, 4))
```
